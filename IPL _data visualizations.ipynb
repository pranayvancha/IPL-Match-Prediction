```python
data = pd.read_csv(r'F:\ASU\IEE_577\Final_Project/matches.csv')
```

```python
import seaborn as sns
fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
sns.countplot(data['venue'],order=data['venue'].value_counts().index,palette='Set2')
plt.xticks(rotation=90,fontsize=11.5)
plt.yticks(fontsize=16)
plt.xlabel('Stadium',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('No. of games hosted in each stadium',fontsize=15)
count=0
venues=pd.DataFrame(data['venue'].value_counts())
venues['name']=data['venue'].value_counts().index
for i in venues['venue']:
    plt.text(count-0.2,i-2,str(i),rotation=90,color='black',size=12)
    count+=1
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
wins=pd.DataFrame(data['winner'].value_counts())
wins['name']=wins.index
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=16)
plt.bar(wins['name'],
        wins['winner'],
        color=['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25','#E82865']
        ,alpha=0.8)
count=0
for i in wins['winner']:
    plt.text(count-0.15,i-4,str(i),size=25,color='black',rotation=90)
    count+=1
plt.title('Total wins by each team',fontsize=30, color= 'red')
plt.xlabel('Teams',fontsize=25)
plt.ylabel('Total no. of matches won',fontsize=24, )
plt.show()
```

```python
fig=plt.gcf()
fig.set_size_inches(18.5,9)
sns.countplot(data['city'],order=data['city'].value_counts().index,palette='Set2')
plt.xticks(rotation=90,fontsize=17)
plt.yticks(fontsize=16)
plt.xlabel('City',fontsize=30)
plt.ylabel('Match Count',fontsize=30)
plt.title('No. of games hosted in each city',fontsize=30, color= 'red')
count=0
cities=pd.DataFrame(data['city'].value_counts())
cities['name']=data['city'].value_counts().index
for i in cities['city']:
    plt.text(count-0.2,i-2,str(i),rotation=90,color='black',size=17)
    count+=1
plt.show()
```

```python
import matplotlib.pyplot as mlt
mlt.style.use('fivethirtyeight')
df_fil=data[data['toss_winner']==data['winner']]
slices=[len(df_fil),(577-len(df_fil))]
mlt.pie(slices,labels=['Win the Toss & win the match','Win the Toss & lose the match'],startangle=90,shadow=True,explode=(0,0),autopct='%1.1f%%',colors=['r','g'])
fig = mlt.gcf()
fig.set_size_inches(6,6)
mlt.show()
```
